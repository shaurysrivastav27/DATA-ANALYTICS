```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
machineutils = pd.read_csv('/mnt/d/pvsc/data.sci/superds/macuti.csv')
machineutils['utils']= 1- machineutils['Percent Idle']  
machineutils.head(10)
```

```python
machineutils['Timestamp'] = pd.to_datetime(machineutils['Timestamp'], format='%d/%m/%Y %H:%M')
```

```python
machineutils.head()
```

```python
dataf = machineutils.pivot_table(columns='Machine',index='Timestamp',values='utils')
dataf.head()
```
